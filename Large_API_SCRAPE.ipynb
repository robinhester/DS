```
from hackernews import HackerNews
import csv
import sqlite3
import pandas
```

```
"""
Setting our HackerNews api to a variable and retrieving
the top 50 stories ids, then getting all of the 'kid' ids
which are comments on the stories. With this we will 
be able to pull all information needed for our database.
"""
hn = HackerNews()
top_stories = hn.top_stories(limit=500)
# column one in story csv
story_ids = str(list(top_stories)).split()

stories = []
for word in story_ids:
    if word.isdigit():
        stories.append(int(word))

# comment ids column one in comment csv
comments = []
for x in stories:
    comments.append(hn.get_item(x).kids)

```

```
"""
Getting the story object by story id then allows to pull
the HackerNews Score, title, and url. Then loading that into
a table in sqlite3, as well as saving a csv version of that table
"""

# story object is required by api to call methods
story_object = hn.get_items_by_ids(stories)

# HackernewsScore of the Story
story_score = []
for index in range(len(stories)):
    story_score.insert(index, story_object[index].score)

# title of Hackernews Story
story_title = []
for index in range(len(stories)):
    story_title.insert(index, story_object[index].title)

# url of Hackernews Story
story_url = []
for index in range(len(stories)):
    story_url.insert(index, story_object[index].url)

# Combining the four lists, url, title, score, and id, 
# into a single tuple to load into a csv files
rows = zip(stories, story_score, story_title, story_url)

# loading tuple into csv file
newfilePath = r"C:/Users/angel/Desktop/storydata.csv"
with open(newfilePath, "w") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)


# set up connection
conn = sqlite3.connect('data.sqlite3')
cursor = conn.cursor()

# creating table stories
create_table = """
                CREATE TABLE IF NOT EXISTS Stories
                (
                    Story_ID int primary key,
                    HackernewsScore int,
                    Title varchar,
                    URL blob
                )
               """

cursor.execute(create_table).fetchall()

# using function to convert a csv to tuple(with encoding)
# cannot load into db without the encoding that is included 
# in this function
csvname = 'largestorydata.csv'
def load_from_csv(csvname):
    """loads a csv as a list of tuples, drops the header."""
    with open(csvname, encoding='utf-8') as f:
        data = [tuple(line) for line in csv.reader(f)]
    data = data[1:]
    return data
data = load_from_csv(csvname)

# inserting data into table of sqlite3 db
insert_query = "INSERT INTO Stories (Story_ID, HackernewsScore, Title, URL) VALUES (?,?,?,?)"
cursor.executemany(insert_query, data)

#commit and close connection
conn.commit()
conn.close()
```
